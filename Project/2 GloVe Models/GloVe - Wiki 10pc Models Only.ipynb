```python
def coo_to_dod(coo, mark = 0):
    # Convert scipy sparse co-occurrence matrix (CSR format) to dict of dicts
    coo_dict_final = {}
    _done = 0
    
    for i, v in coo.todok().items():
        if i[0] not in coo_dict_final:
            coo_dict_final[i[0]] = {}
        coo_dict_final[i[0]].update({i[1] : v})
        
        _done += 1
        if mark > 0 and _done % mark == 0:
            print('  Processed {} items'.format(_done))
    
    return coo_dict_final
```

```python
def invert_dict(dic):
    return {value:key for (key, value) in dic.items()}
```

```python
%%time

batch_dfs = {}

for bb in range(batch_count):
    print('Processing batch {} of {}'.format(bb+1,batch_count))
    
    print(' Unpickling matrix')
    with open(datapath+'/Corpora/wiki/enwiki_20200520/Pickles/10pc_b{}_coo.pkl'.format(bb+1), 'rb') as pfile:
        coo = pickle.load(pfile)
    
    print(' Converting to COOrdinate format')
    coo = coo.asformat('coo')
    
    
    print(' Building GloVe model')
    model = Glove(no_components=300, 
                  alpha = 0.75, 
                  max_count = 100,
                  learning_rate=0.05)
    
    model.fit(coo, epochs=25, no_threads=8, verbose=True)
    
    print(' Adding dictionary')
    
    #  Load pickled dictionary
    with open(datapath+'/Corpora/wiki/enwiki_20200520/Pickles/10pc_b{}_dict.pkl'.format(bb+1), 'rb') as pfile:
        batch_dict = pickle.load(pfile)
       
    model.add_dictionary(invert_dict(batch_dict))
    
    # Save model
    print(' Saving model')
    model.save(datapath+'/Models/2 GloVe/wiki10pc_glove_batch{}.model'.format(bb+1))   

```
