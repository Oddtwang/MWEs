```python
import pickle
from gensim import corpora
from scipy import sparse
import itertools

from dist_cooccurrence import dist_cooccurrence_v3
```

```python
%%time

# Iterate over batches
for bb in range(9,batch_count):

    print('Executing for batch {}'.format(bb+1))
    
    this_batch = ngram_eval[ngram_eval.batch == bb+1].reset_index(drop=True)

    #  Load pickled vocab set
    with open(datapath+'/Corpora/wiki/enwiki_20200520/10pc_b{}vocab.pkl'.format(bb+1), 'rb') as pfile:
        batch_vocab = pickle.load(pfile)
        
    # Dictionary mapping
    batch_dict = corpora.Dictionary([['<unk>']])
    batch_dict.add_documents([list(batch_vocab)])
    
    # Pickle and save dictionary
    with open(datapath+'/Corpora/wiki/enwiki_20200520/Pickles/10pc_b{}_dict.pkl'.format(bb+1), 'wb') as pfile:
        pickle.dump(batch_dict, pfile)
    
    
    # Distributed construction of matrix
    cooccurrence = dist_cooccurrence_v3(batch_dict, 
                                     sent_gen(w10p, vocab=batch_vocab), 
                                     window_size = 20,
                                        mark=100000,
                                        rem_return = 50000,
                                    specs=[('popen', 6)])
    
    # Pickle and save
    with open(datapath+'/Corpora/wiki/enwiki_20200520/Pickles/10pc_b{}_coo.pkl'.format(bb+1), 'wb') as pfile:
        pickle.dump(cooccurrence, pfile, protocol=4)
        
    print('Pickled')
    print('****************************','\n')
    
    del cooccurrence
    
```
