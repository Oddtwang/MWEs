```python
def invert_dict(dic,shift=0):
    return {value:key+shift for (key, value) in dic.items()}

def removekey(d, key):
    r = dict(d)
    del r[key]
    return r
```

```python
batch_count = 1
```

```python
%%time

batch_dfs = {}

for bb in range(batch_count):
    print('Processing batch {} of {}'.format(bb+1,batch_count))
    
    print(' Unpickling matrix')
    with open(datapath+'/Corpora/wiki/simple_20200601/Pickles/simp_b{}_coo.pkl'.format(bb+1), 'rb') as pfile:
        coo = pickle.load(pfile)
    
    print(' Converting to COOrdinate format')
    # Drop 0th row and column, corresponding to <unk> token
    coo = coo[1:,1:].asformat('coo')
    
    
    print(' Building GloVe model')
    model = Glove(no_components=300, 
                  alpha = 0.75, 
                  max_count = 100,
                  learning_rate=0.05)
    
    model.fit(coo, epochs=25, no_threads=8, verbose=True)
    
    print(' Adding dictionary')
    
    #  Load pickled dictionary
    with open(datapath+'/Corpora/wiki/simple_20200601/Pickles/simp_b{}_dict.pkl'.format(bb+1), 'rb') as pfile:
        batch_dict = pickle.load(pfile)
       
    batch_dict = removekey(batch_dict, 0)
    
    model.add_dictionary(invert_dict(batch_dict))
    
    # Save model
    print(' Saving model')
    model.save(datapath+'/Models/2 GloVe/simp_glove_coo_v2_batch{}.model'.format(bb+1))   

```
