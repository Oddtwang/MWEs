```python
from gensim.corpora import WikiCorpus, MmCorpus, Dictionary
from gensim import utils
from numpy import random
```

```python
from mywikicorpus import MyWikiCorpus
```

```python
wiki = MyWikiCorpus(datapath+'/Corpora/wiki/enwiki-20200520-pages-articles-multistream.xml.bz2',
                  lower=False,
                  token_min_len = 1,
                  token_max_len = 40
                 )
```

```python
output = open(datapath+'/Corpora/wiki/enwiki_20200520/enwiki_20200520_10pc.txt', 'w', encoding='utf-8')

i = 0
sample_proportion = 0.10

for text in wiki.get_texts():
    if random.random_sample() < sample_proportion:
        output.write(bytes(' '.join(text), 'utf-8').decode('utf-8')+'\n')
        
    i = i + 1
    if (i % 10000 == 0):
        print('Processed ' + str(i) + ' articles')
        
output.close()
print('Processing complete!')
```
