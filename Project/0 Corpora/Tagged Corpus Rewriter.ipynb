```python
import re
```

```python
_pos = re.compile("\s+'s\|POS\s+")
_pos2 = re.compile("\s+'\|POS\s+")
_posopen = re.compile("^(\s+)?'s?\|POS\s+")

_hyph = re.compile("\s+\-\|HYPH\s+")
_hyph2 = re.compile("\s+[\\\/]\|HYPH\s+")
_hyphopen= re.compile("^(\s+)?[\-\\\/]\|HYPH\s+")
_hyphend= re.compile("\s+[\-\\\/]\|HYPH(\s+)?$")

_hyphsym = re.compile("\s+-\|HYPH\s+–\|SYM\s+")
_hyphsymo = re.compile("^(\s+)?-\|HYPH\s+–\|SYM\s+")
_hyphsyme = re.compile("\s+-\|HYPH\s+–\|SYM(\s+)$")

_perc = re.compile("\|CD\s+%\|NN\s+")
_percopen = re.compile("^(\s+)?%\|NN\s+")

_range = re.compile("([\d\.,]+)\|CD\s+\–\|SYM\s+([\d\.,]+)\|CD\s+")
_neg = re.compile("\s+\–\|SYM\s+([\d\.,]+)\|CD\s+")

def tag_cleaner(line):
    l = re.sub(_posopen, '', line)

    l = re.sub(_pos, "'s ",l)
    l = re.sub(_pos2, "' ",l)
    
    l = re.sub(_hyphsymo, '', l)
    l = re.sub(_hyphsyme, ' ', l)
    l = re.sub(_hyphsym, ' –|SYM ', l)
    
    l = re.sub(_hyphopen, '', l)
    l = re.sub(_hyphend, ' ', l)
    l = re.sub(_hyph, '-', l)
    l = re.sub(_hyph2, '/', l)
    
    l = re.sub(_perc, '%|CD ', l)
    l = re.sub(_percopen, '', l)

    l = re.sub(_range,r"\1–\2|CD ", l)
    l = re.sub(_neg, r" –\1|CD ", l)

    return l
```

```python
testl = "'s|POS %|NN born|VBN in|IN Canada|NNP 52.0|CD %|NN in|IN the|DT 2007|CD –|SYM 08|CD season|NN "
testl = ' Timothy|NNP I|PRP 727|CD –|SYM 823|CD wrote|VBD of|IN'
testl = 'the|DT bookness|NN of|IN books|NNS -|HYPH –|SYM how|WRB'
testl = ' Timothy|NNP I|PRP  –|SYM 823.45|CD wrote|VBD of|IN'

tag_cleaner(testl)
```

```python
for l in tagfile.readlines():
    print(tag_cleaner(l.decode('utf-8')))
```

```python
tagfile = open(datapath+'/Corpora/wiki/simple_20200601/simple_sample_tagged_3w.txt','rb')
output = open(datapath+'/Corpora/wiki/simple_20200601/simple_sample_tagged_3w_clean.txt', 'w', encoding='utf-8')

i = 0
for l in tagfile.readlines():
    output.write(tag_cleaner(l.decode('utf-8')))
    i = i + 1
    if (i % 100000 == 0):
        print('Processed ' + str(i) + ' sentences')
        
tagfile.close()
output.close()
print('Processing complete!')
```
