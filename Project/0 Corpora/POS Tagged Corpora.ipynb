```python
import pickle
import pandas as pd
import re
import numpy as np

import nltk
from nltk.tokenize import MWETokenizer

from nltk.corpus.reader.plaintext import PlaintextCorpusReader
    
from nltk.tokenize import WhitespaceTokenizer

from generators import sent_gen, word_gen, Sent_Seq
```

```python
tagger = POSClient(host='localhost', port=9198)
```

```python
samp = PlaintextCorpusReader(datapath+'/Corpora/wiki/simple_20200601/','simple_sample.txt',
                            word_tokenizer = WhitespaceTokenizer()
                            )
```

```python
def tup_to_str(tup, tagmark='|', unk='<unk>'):
    if tup[0] == unk:
        return tup[0]
    else:
        return tup[0]+tagmark+tup[1]
```

```python
for sent in sent_gen(samp, asstr=True):
    sent = [ tup_to_str(tup) for tup in tagger.tag(sent)]
    print(sent)
```

```python
output = open(datapath+'/Corpora/wiki/simple_20200601/simple_sample_tagged.txt', 'w', encoding='utf-8')

i = 0
for sent in sent_gen(samp, tagger=tagger):
    output.write(bytes(' '.join(sent), 'utf-8').decode('utf-8')+' \n')
    i = i + 1
    if (i % 100000 == 0):
        print('Processed ' + str(i) + ' sentences')
        
output.close()
print('Processing complete!')
```
